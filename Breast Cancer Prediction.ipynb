```python
cols_to_norm=['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
```

```python
df[cols_to_norm]=df[cols_to_norm].apply(lambda x: (x-x.min()) / (x.max()-x.min()))
```

```python
a=tf.feature_column.numeric_column('radius_mean')
b=tf.feature_column.numeric_column('texture_mean')
c=tf.feature_column.numeric_column('perimeter_mean')
d=tf.feature_column.numeric_column('area_mean')
e=tf.feature_column.numeric_column('smoothness_mean')
f=tf.feature_column.numeric_column('compactness_mean')
g=tf.feature_column.numeric_column('concavity_mean')
h=tf.feature_column.numeric_column('radius_se')
i=tf.feature_column.numeric_column('texture_se')
j=tf.feature_column.numeric_column('perimeter_se')
k=tf.feature_column.numeric_column('area_se')
l=tf.feature_column.numeric_column('smoothness_se')
m=tf.feature_column.numeric_column('compactness_se')
n=tf.feature_column.numeric_column('concavity_se')
o=tf.feature_column.numeric_column('concave points_se')
p=tf.feature_column.numeric_column('symmetry_se')
q=tf.feature_column.numeric_column('fractal_dimension_se')
r=tf.feature_column.numeric_column('radius_worst')
s=tf.feature_column.numeric_column('texture_worst')
t=tf.feature_column.numeric_column('perimeter_worst')
u=tf.feature_column.numeric_column('area_worst')
v=tf.feature_column.numeric_column('smoothness_worst')
w=tf.feature_column.numeric_column('compactness_worst')
x=tf.feature_column.numeric_column('concavity_worst')
y=tf.feature_column.numeric_column('concave points_worst')
z=tf.feature_column.numeric_column('symmetry_worst')
ab=tf.feature_column.numeric_column('fractal_dimension_worst')
```

```python
feat_cols=[a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,q,r,s,t,u,v,w,x,y,z,ab]
```

```python
x_data=df.drop(['diagnosis', 'Unnamed: 32','id'],axis=1)
```

```python
x_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['diagnosis']= le.fit_transform(df['diagnosis'])
```

```python
labels=df['diagnosis']
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x_data,labels,test_size=0.3,random_state=101)
```

```python
input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model=tf.compat.v1.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
```

```python
model.train(input_fn=input_func,steps=1000)
```

```python
eval_input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_test,y=y_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
results=model.evaluate(eval_input_func)
```

```python
results
```

```python
pred_input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
predictions=model.predict(pred_input_func)
```
